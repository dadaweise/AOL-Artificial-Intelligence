```python
import os
import pandas as pd

def data_dictionary():
    # 1. Pastikan jalur folder sudah benar
    path_train = "Data/train/"
    path_test = "Data/test/"

    # 2. Inisialisasi variabel yang dibutuhkan
    if not os.path.exists(path_train):
        print(f"Error: Folder {path_train} tidak ditemukan!")
        return None

    list_train = os.listdir(path_train)
    list_test = os.listdir(path_test) if os.path.exists(path_test) else []

    train_dictionary = {"image_path": [], "target": []}

    # 3. Loop untuk Data Train
    k = 0
    for i in list_train:
        path_disease_train = os.path.join(path_train, i)

        # Pastikan yang dibaca adalah folder, bukan file tersembunyi
        if not os.path.isdir(path_disease_train):
            continue

        image_list_train = os.listdir(path_disease_train)
        for j in image_list_train:
            img_path_train = os.path.join(path_disease_train, j)

            # Filter file bermasalah seperti saat di Kaggle
            if "Normal/34.avif" not in img_path_train.replace("\\", "/"):
                train_dictionary["image_path"].append(img_path_train)
                train_dictionary['target'].append(k)
        k += 1

    # 4. Loop untuk Data Test
    for i in list_test:
        path_disease_test = os.path.join(path_test, i)
        if not os.path.isdir(path_disease_test):
            continue

        image_list_test = os.listdir(path_disease_test)
        for j in image_list_test:
            img_path_test = os.path.join(path_disease_test, j)
            train_dictionary["image_path"].append(img_path_test)
            train_dictionary['target'].append(4) # Label target khusus test

    train_df = pd.DataFrame(train_dictionary)
    return train_df
```

```python
import tensorflow as tf
import numpy as np
import cv2

import tensorflow as tf
import numpy as np
import cv2

def predict_skin_disease(image_path):
    # 1. Load model yang sudah di-save ulang tadi
    # Pastikan path filenya benar
    model = tf.keras.models.load_model('skin_model_fixed.keras')

    # 2. Kita butuh VGG16 lagi untuk ekstraksi fitur
    from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
    vgg_base = VGG16(weights='imagenet', include_top=False, input_shape=(180, 180, 3))

    # 3. Load & Preprocess Gambar
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # OpenCV pake BGR, Keras pake RGB
    img = cv2.resize(img, (180, 180))
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img) # Pakai fungsi bawaan VGG agar lebih akurat

    # 4. Ekstraksi Fitur (Harus sama dengan saat training)
    features = vgg_base.predict(img)
    features = features.reshape(1, -1) # Akan jadi 12800 fitur

    # 5. Prediksi Akhir
    pred = model.predict(features)[0]
    predicted_class = np.argmax(pred)

    return predicted_class
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input

# 1. Pastikan vgg_model didefinisikan dengan input_shape yang jelas
# Jika vgg_model sudah ada di memori, baris ini akan memastikan dimensinya terkunci
vgg_model = VGG16(weights='imagenet', include_top=False, input_shape=(180, 180, 3))

def extract_features_safely(data):
    # Paksa konversi ke numpy array dan pastikan ukurannya (N, 180, 180, 3)
    data_array = np.array(data)

    # Jika data_array berbentuk (N,) yang berisi list, kita harus menumpuknya manual
    if len(data_array.shape) == 1:
        data_array = np.stack(data)

    print(f"Bentuk data terdeteksi: {data_array.shape}")

    # Preprocessing resmi VGG16
    data_preprocessed = preprocess_input(data_array.astype('float32'))

    # Gunakan tf.convert_to_tensor untuk menghindari error 'unknown TensorShape'
    data_tensor = tf.convert_to_tensor(data_preprocessed)

    # Ekstraksi fitur
    features = vgg_model.predict(data_tensor, batch_size=32, verbose=1)

    # Flatten hasil (dari 4D ke 2D: 12800 fitur)
    return features.reshape(features.shape[0], -1)

# --- EKSEKUSI ---
print("Memulai ekstraksi fitur x_train...")
x_train_features = extract_features_safely(x_train)

print("\nMemulai ekstraksi fitur x_test...")
x_test_features = extract_features_safely(x_test)

print(f"\nSelesai! Fitur x_train sekarang: {x_train_features.shape}")
```

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.optimizers import Adam

# 1. Deteksi jumlah kelas otomatis (Ingat: tadi ada label sampai indeks 19)
# 1. Pastikan jumlah kategori sesuai dengan data kamu (tadi terdeteksi sampai 20)
import numpy as np
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam

# 1. Sesuaikan num_classes dengan data target kamu (23)
num_classes = 23

# 2. Rakit ulang model dengan 23 output neurons
model_final = models.Sequential([
    layers.Input(shape=(12800,)),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax') # Sekarang slotnya sudah pas 23
])

model_final.compile(
    optimizer=Adam(learning_rate=0.00005),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Memulai training... Dimensi kategori sudah sinkron!")

early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
# 3. Jalankan lagi - Gunakan batch_size besar agar lebih efisien
model_final.fit(x_train, y_train_fixed,
                epochs=50,
                batch_size=256,
                validation_data=(x_test, y_test_fixed),
                verbose=1)  

# 4. Simpan model yang sudah sinkron
model_final.save('skin_model_v2_fix_23.keras')

# 5. Fungsi Prediksi Langsung

def quick_predict(image_path):
    # 1. WAJIB: Pastikan list ini punya 23 nama sesuai urutan folder di 'Data/train'
    # Kamu bisa cek urutannya dengan: sorted(os.listdir('Data/train'))
    class_names = [
        "Acne", "Actinic Keratosis", "Atopic Dermatitis", "Bullous Disease",
        "Cellulitis", "Eczema", "Exanthems", "Hair Loss", "Herpes",
        "Light Diseases", "Lupus", "Melanoma", "Nail Fungus", "Poison Ivy",
        "Psoriasis", "Scabies", "Seborrheic Keratoses", "Systemic Disease",
        "Tinea", "Urticaria Hives", "Vascular Tumors", "Vasculitis", "Warts"
    ]

    if not os.path.exists(image_path):
        return "Error: File tidak ditemukan!"

    # 2. Load model yang baru saja kamu latih (v2_fix_23)
    test_model = tf.keras.models.load_model('skin_model_v2_fix_23.keras')

    # 3. Preprocess Gambar
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (180, 180))

    # --- PENTING: Gunakan Preprocessing Resmi VGG16 ---
    img_array = np.expand_dims(img, axis=0)
    img_preprocessed = preprocess_input(img_array.astype('float32'))

    # 4. Ekstraksi Fitur
    features = vgg_model.predict(img_preprocessed)
    features = features.reshape(1, -1)

    # 5. Hasil Akhir
    pred = test_model.predict(features)[0]
    predicted_idx = np.argmax(pred)

    # Proteksi jika indeks melebihi jumlah nama yang tersedia
    if predicted_idx >= len(class_names):
        return f"Indeks {predicted_idx} terdeteksi, tapi nama tidak ada di list!"

    return class_names[predicted_idx]

# TEST
path_gambar = r"Data\test\Acne and Rosacea Photos\07sebDerem1101051.jpg"
print(f"\nHasil Diagnosa SkinSpectra: {quick_predict(path_gambar)}")
```
